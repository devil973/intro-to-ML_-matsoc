#### Import the necessary library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Read the data set and check the name of columns data set, print the first five row and and details of the data set using .info() & .describe()

```python
df=pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

#### Check if there are any null values

```python
#this prints the total null values in our dataframe df 
null_count = df.isnull().sum().sum()
print(f"Total null values: {null_count}")
```

## Train Test and Split

### Import train test split 

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,0:30]
y = df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

## Decision Tress

#### Import Decision Tree Classifier 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

#### Fit the model

```python
dtree.fit(X_train,y_train)
```

## Prediction and Evaluation

```python
predictions =dtree.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn import tree
plt.figure(figsize=(20,25))
feature_names = list(X.columns)
tree.plot_tree(dtree,feature_names=feature_names,class_names=['Class-1', 'Class-0'],rounded=True, # Rounded node edges
          filled=True, # Adds color according to class
          proportion=True
        )
plt.show()
```
